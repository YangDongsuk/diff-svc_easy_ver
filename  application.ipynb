```python
#@title #3-2. wavíŒŒì¼ +-12 ìŒì • ë”í•´ì£¼ê¸°

#@markdown ### **âœ¨ Augmentation**
#@markdown ì´ì œ ì•„ê¹Œ 15ì´ˆ ë‹¨ìœ„ë¡œ ì˜ë¼ì§„ ìŒì„±ì— +12, -12 ìŒì •ì„ ì¶”ê°€ í•´ì¤„ ê²ƒì…ë‹ˆë‹¤.

#@markdown #### **â­ï¸ ê·¸ëƒ¥ ëŒë¦¬ë©´ ê³ ìŒ ì €ìŒ ë¶€ë¶„ì´ ê¹¨ì§€ê¸° ë•Œë¬¸ì— Augmentationì˜ ì¼ì¢…ìœ¼ë¡œ ê³ ìŒ ì €ìŒ ë°ì´í„°ë¥¼ ì¶”ê°€í•´ì£¼ëŠ” ê²ƒì…ë‹ˆë‹¤.**
#@markdown ì´˜ì´˜í•˜ê²Œ Augmentationì„ ì§„í–‰í•˜ëŠ” ê²ƒë³´ë‹¤ +12, -12ë§Œ í•´ì£¼ì—ˆì„ ë•Œ ê°€ì¥ íš¨ê³¼ì ì´ì—ˆìœ¼ë¯€ë¡œ ìë™ìœ¼ë¡œ ì´ë ‡ê²Œ ì²˜ë¦¬í•˜ë„ë¡ í•˜ì˜€ìŠµë‹ˆë‹¤. \
#@markdown í•„ìˆ˜ì ì¸ ê³¼ì •ì€ ì•„ë‹ˆê¸°ì— ì›ì¹˜ ì•Šìœ¼ì‹œë©´ ë„˜ì–´ê°€ì…”ë„ ì¢‹ìœ¼ë‚˜, ì„±ëŠ¥ì„ ìœ„í•´ ì§„í–‰í•˜ì‹œëŠ” ê²ƒì„ ê¶Œì¥í•©ë‹ˆë‹¤.

#@markdown âš¡ï¸ ì‹œê°„ì´ ë‹¤ì†Œ ì˜¤ë˜ ê±¸ë¦½ë‹ˆë‹¤.

#@markdown ---

#@markdown ê²°ê³¼ë¬¼ì€ ì•„ë˜ ê²½ë¡œì— ì €ì¥ë©ë‹ˆë‹¤. \
#@markdown /content/drive/MyDrive/ì„¤ì •í•œ í”„ë¡œì íŠ¸ ì´ë¦„(singer_name)/wav_cut_add_pitches

#@markdown \
#@markdown #### [ì¬êµìœ¡ì‹œğŸ‘¾] ì‹¤í–‰í•˜ì§€ ì•Šìœ¼ì‹œë©´ ë©ë‹ˆë‹¤.


import os
import glob
import scipy.io.wavfile as wav
import numpy as np
import pyworld as world
import concurrent.futures
import logging

# ë¡œê·¸ ì„¤ì •
logging.basicConfig(format='%(asctime)s: %(message)s', level=logging.INFO, datefmt='%x %a %X')

# ìŒì • ë³€í™˜ ë²”ìœ„ ì„¤ì • (+-12 ë° ì›ë³¸)
pitches = [-12, 0, 12]

def shift_pitch(path):
    loc, _ = os.path.splitext(path)
    directory, fname = os.path.split(loc)

    # ì¶œë ¥ ë””ë ‰í† ë¦¬ ì„¤ì • ë° ìƒì„±
    output_dir = f'/content/drive/MyDrive/{singer_name}/wav_cut_add_pitches'
    os.makedirs(output_dir, exist_ok=True)

    logging.info(f'Loading {fname}')
    fs, x = wav.read(path)
    xtype = x.dtype
    int_type = np.issubdtype(xtype, np.integer)

    if int_type:
        info = np.iinfo(xtype)
        x = x / info.max

    if len(x.shape) == 2:
        x = (x[:,0] + x[:,1]) / 2

    logging.info(f'Analyzing {fname}')
    f0, t = world.harvest(x, fs)
    sp = world.cheaptrick(x, f0, t, fs)
    ap = world.d4c(x, f0, t, fs, threshold=0.25)

    for i in pitches:
        logging.info(f'Synthesizing {fname} {i:+}')
        shift_f0 = f0 * np.exp2(i / 12)
        y = world.synthesize(shift_f0, sp, ap, fs)

        if int_type:
            y = (y * info.max).clip(info.min, info.max).astype(xtype)

        # ë³€í™˜ëœ íŒŒì¼ ì €ì¥
        wav.write(os.path.join(output_dir, f'{fname}{i:+}.wav'), fs, y)

# ë©”ì¸ í•¨ìˆ˜ ì‹¤í–‰
if __name__ == '__main__':
    logging.info('Starting process pool')
    
    # ë³€í™˜ ëŒ€ìƒ íŒŒì¼ë“¤ì´ ìˆëŠ” ë””ë ‰í† ë¦¬ ì§€ì •
    input_dir =  f'/content/drive/MyDrive/{singer_name}/wav_cut'
    
    with concurrent.futures.ProcessPoolExecutor() as executor:
        # ì§€ì •ëœ ë””ë ‰í† ë¦¬ì˜ ëª¨ë“  wav íŒŒì¼ì— ëŒ€í•´ ìŒì • ë³€í™˜ ì‹¤í–‰
        executor.map(shift_pitch, glob.glob(os.path.join(input_dir, '*.wav')))
```
