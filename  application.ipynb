```python
#@title #3-1-2.  wav파일 15초 단위로 자르기

from pydub import AudioSegment
import os

#@markdown ### **✨ 데이터셋 파일의 경로 지정하기**

#@markdown 저는 구글드라이브 최상위 경로에 ssmovie.wav라는 이름으로 되어있어서 다음과 같은 경로입니다. \
#@markdown 경로는 왼쪽의 폴더이미지 눌러서 파일을 찾아서 그 왼쪽에 점 3개눌러서 경로 복사로 붙여넣기하면 됩니다. \
#@markdown 구글 드라이브 이용할테니 drive폴더 -> MyDrive -> wav파일을 업로드한 위치에서 찾습니다.
input_filepath = '/content/drive/MyDrive/ssmovie.wav' #@param {type: "string"}
#@markdown #### *🔥 제출된 dataset 폴더의 talking_data를 활용하시면 됩니다*

#@markdown \
#@markdown #### **⭐️ 데이터셋의 길이에 따라 아래와 같은 차이점이 나타납니다.**
#@markdown 1️⃣ 60분: 안정적으로 결과물을 얻어낼 수 있습니다. \
#@markdown 2️⃣ 20분: Step 수를 증가시켜도 결과물의 퀄리티가 크게 좋아지지 못합니다. \
#@markdown 3️⃣ 5분: MCD 값을 신뢰할 수 없게 되므로 완벽한 퀄리티를 갖추기 어렵습니다. 

#@markdown \
#@markdown ### **✨ wav파일 15초 단위로 자르기**
#@markdown 당신이 올린 음성파일은 15초씩 잘라진 채로 \
#@markdown /content/drive/MyDrive/설정한 프로젝트 이름(singer_name)/wav_cut \
#@markdown 에 저장됩니다. \


#@markdown 이미 5~15초 사이로 쪼개진 파일들을 가지고 계신 경우에는, \
#@markdown 위 경로에 넣어두시면 됩니다. \\

#@markdown \\
#@markdown #### [재교육시👾] 실행하실 필요 없습니다.


# singer_name = 'test1'  # 가수 이름 지정

# 출력 디렉토리 설정 및 생성
output_dir = f'/content/drive/MyDrive/{singer_name}/wav_cut'
os.makedirs(output_dir, exist_ok=True)

# 파일 읽기
audio = AudioSegment.from_wav(input_filepath)

# 15초 단위로 분할
duration_ms = len(audio)
interval = 15 * 1000  # 15초를 밀리초로 변환
start = 0
part = 0

while start < duration_ms:
    end = min(start + interval, duration_ms)
    segment = audio[start:end]
    
    # 분할된 파일 저장
    output_filename = os.path.join(output_dir, f'{os.path.splitext(os.path.basename(input_filepath))[0]}_{part}.wav')
    segment.export(output_filename, format='wav')
    
    part += 1
    start = end

print(f'Files saved in {output_dir}')
```

```python
#@title #3-2. wav파일 +-12 음정 더해주기

#@markdown ### **✨ Augmentation**
#@markdown 이제 아까 15초 단위로 잘라진 음성에 +12, -12 음정을 추가 해줄 것입니다.

#@markdown #### **⭐️ 그냥 돌리면 고음 저음 부분이 깨지기 때문에 Augmentation의 일종으로 고음 저음 데이터를 추가해주는 것입니다.**
#@markdown 촘촘하게 Augmentation을 진행하는 것보다 +12, -12만 해주었을 때 가장 효과적이었으므로 자동으로 이렇게 처리하도록 하였습니다. \
#@markdown 필수적인 과정은 아니기에 원치 않으시면 넘어가셔도 좋으나, 성능을 위해 진행하시는 것을 권장합니다.

#@markdown ⚡️ 시간이 다소 오래 걸립니다.

#@markdown ---

#@markdown 결과물은 아래 경로에 저장됩니다. \
#@markdown /content/drive/MyDrive/설정한 프로젝트 이름(singer_name)/wav_cut_add_pitches

#@markdown \
#@markdown #### [재교육시👾] 실행하지 않으시면 됩니다.


import os
import glob
import scipy.io.wavfile as wav
import numpy as np
import pyworld as world
import concurrent.futures
import logging

# 로그 설정
logging.basicConfig(format='%(asctime)s: %(message)s', level=logging.INFO, datefmt='%x %a %X')

# 음정 변환 범위 설정 (+-12 및 원본)
pitches = [-12, 0, 12]

def shift_pitch(path):
    loc, _ = os.path.splitext(path)
    directory, fname = os.path.split(loc)

    # 출력 디렉토리 설정 및 생성
    output_dir = f'/content/drive/MyDrive/{singer_name}/wav_cut_add_pitches'
    os.makedirs(output_dir, exist_ok=True)

    logging.info(f'Loading {fname}')
    fs, x = wav.read(path)
    xtype = x.dtype
    int_type = np.issubdtype(xtype, np.integer)

    if int_type:
        info = np.iinfo(xtype)
        x = x / info.max

    if len(x.shape) == 2:
        x = (x[:,0] + x[:,1]) / 2

    logging.info(f'Analyzing {fname}')
    f0, t = world.harvest(x, fs)
    sp = world.cheaptrick(x, f0, t, fs)
    ap = world.d4c(x, f0, t, fs, threshold=0.25)

    for i in pitches:
        logging.info(f'Synthesizing {fname} {i:+}')
        shift_f0 = f0 * np.exp2(i / 12)
        y = world.synthesize(shift_f0, sp, ap, fs)

        if int_type:
            y = (y * info.max).clip(info.min, info.max).astype(xtype)

        # 변환된 파일 저장
        wav.write(os.path.join(output_dir, f'{fname}{i:+}.wav'), fs, y)

# 메인 함수 실행
if __name__ == '__main__':
    logging.info('Starting process pool')
    
    # 변환 대상 파일들이 있는 디렉토리 지정
    input_dir =  f'/content/drive/MyDrive/{singer_name}/wav_cut'
    
    with concurrent.futures.ProcessPoolExecutor() as executor:
        # 지정된 디렉토리의 모든 wav 파일에 대해 음정 변환 실행
        executor.map(shift_pitch, glob.glob(os.path.join(input_dir, '*.wav')))
```

```python

#@title #7-2. 결과물 출력
#@markdown ### **✨ mr 제거 wav파일 불러오기**

#@markdown 자, 이제 원하는 노래의 mr제거 버전 wav파일들을 넣어주세요. \
#@markdown 변환하고 싶은 음악의 mr이 제거된 보컬 음성 파일을 wav 형식으로 준비하셔야 합니다.  \

import os

# 입력 폴더 경로
#@markdown 저는 /content/drive/MyDrive/no_mr_small 폴더 안에 타겟 보컬 음성 wav파일들을 넣었습니다. \
#@markdown 경로는 왼쪽의 폴더이미지 눌러서 파일을 찾아서 그 왼쪽에 점 3개눌러서 경로 복사로 붙여넣기하면 됩니다.


wav_in_folder = "/content/drive/MyDrive/no_mr_small" #@param {type: "string"}
#@markdown #### *🔥 제출된 dataset 폴더의 target_vocal을 활용하시면 됩니다*

#@markdown 결과물은 ' /content/drive/MyDrive/당신이 설정한 프로젝트 이름(singer_name)/result'에 저장됩니다. \
#@markdown 이름은 ' 원본파일 + 당신이 설정한 프로젝트 이름(singer_name).wav'로 저장됩니다.

#@markdown \
#@markdown #### **⭐️ 일반적으로 타겟 보컬의 음질이 좋을수록, 데이터셋과 pitch가 유사할수록 결과물이 좋게 나옵니다.**
#@markdown 타겟 음원의 mr제거나 pitch 조정 등은 https://vocalremover.org/ko/ 사이트에서 진행하세요!

#@markdown \
#@markdown #### **👻 이제 원하는 목소리로 불러진 노래를 들어보시면 됩니다!**

#@markdown \
#@markdown #### [재교육시👾] 실행하세요. 만약 출력물 중 이름이 중복된 파일이 있으면 없을때까지 _re가 중간에 계속 붙도록 했습니다.


# 모든 wav 파일 가져오기
wav_in_files = [os.path.join(wav_in_folder, file) for file in os.listdir(wav_in_folder) if file.endswith('.wav')]

# 출력 경로 설정
output_folder = f"/content/drive/MyDrive/{singer_name}/result"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)  # 경로가 존재하지 않는 경우 폴더 생성
key = 0


pndm_speedup = 100

add_noise_step = 1000 

thre = 0.05 
use_crepe= True 
use_pe=False 
use_gt_mel= True 
# 각 입력 파일에 대해 처리
# 각 입력 파일에 대해 처리
for wav_in in wav_in_files:
    # 출력 파일 이름 생성 (입력 파일 이름에 프로젝트 이름 추가)
    wav_out_base = os.path.basename(wav_in)  # 입력 파일의 기본 이름 (확장자 포함)
    wav_out_base_no_ext = os.path.splitext(wav_out_base)[0]  # 확장자 제거
    wav_out = f"{wav_out_base_no_ext}_{singer_name}.wav"  # 프로젝트 이름 추가

    # 중복된 파일 이름이 존재할 경우 '_re' 추가
    while os.path.exists(f"{output_folder}/{wav_out}"):
        wav_out_base_no_ext = wav_out_base_no_ext + '_re'
        wav_out = f"{wav_out_base_no_ext}_{singer_name}.wav"

    # 모델 실행 및 결과 저장
    f0_tst, f0_pred, audio = run_clip(svc_model, file_path=wav_in, key=key, acc=pndm_speedup, use_crepe=use_crepe, use_pe=use_pe, thre=thre, use_gt_mel=use_gt_mel, add_noise_step=add_noise_step, project_name=project_name, out_path=f"{output_folder}/{wav_out}")
```
